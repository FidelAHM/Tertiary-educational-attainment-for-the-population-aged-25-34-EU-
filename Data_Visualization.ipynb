```python
import pandas as pd
import numpy as np
```

```python

df = pd.read_csv("estat_tgs00105.tsv", sep="\t", engine="python")
```

```python
df.rename(
    columns={'freq,unit,isced11,age,sex,geo\\TIME_PERIOD': 'meta'},
    inplace=True
)

list(df.columns)
```

```python
df[['freq', 'unit', 'isced11', 'age', 'sex', 'geo']] = (
    df['meta'].str.split(',', expand=True)
)

df.drop(columns=['meta'], inplace=True)

list(df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# Convert from wide  long format (VERY IMPORTANT)
# Eurostat data has years as columns.
# Tableau works best when year is a column.

df_long = df.melt(
    id_vars=['sex', 'age', 'geo'],
    var_name='year',
    value_name='value'
)

df_long.head()
```

```python
# We keep:

# Age: 2534

# Sex: Male (M) and Female (F)
df_long = df_long[
    (df_long['age'] == 'Y25-34') &
    (df_long['sex'].isin(['M', 'F']))
]
```

```python
#Clean Eurostat flags
df_long['value'] = (
    df_long['value']
    .astype(str)
    .str.replace(':', '', regex=False)
    .str.replace('(b)', '', regex=False)
    .str.strip()
)

df_long['value'] = pd.to_numeric(df_long['value'], errors='coerce')
```

```python
df_long.dtypes
```

```python
country_df = (
    df_long
    .groupby(['geo', 'sex', 'year'], as_index=False)['value']
    .mean()
)

country_df.head()
```

```python
# Pivot Male vs Female (required for gender gap)
pivot = country_df.pivot_table(
    index=['geo', 'year'],
    columns='sex',
    values='value'
).reset_index()

pivot.head()
```

```python
# Compute the Gender Gap (Female  Male)
pivot['gender_gap'] = pivot['F'] - pivot['M']
pivot.head()
```

```python
# Sort & check
pivot.sort_values(['year', 'geo']).head()
```

```python
#Export Tableau-ready CSV
pivot.to_csv("tertiary_education_gender_clean.csv", index=False)
```
